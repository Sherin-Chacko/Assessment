```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_path = "avg_rain_state.csv"
df = pd.read_csv(csv_path, low_memory = False)
```

```python
x-axis = np.arange(len(df))
tick_locations = [value for value in x-axis]
plt.figure(figsize=(20,3)
plt.bar(x-axis, vertical)
plt.xticks(tick_locations, df("state"), rotation='horizontal')
plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain(inches)")
plt.show()
```
